```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creates a black image and assign it to a variable.
blank_img = np.zeros(shape=(1000,1000,3), dtype=np.int16)
```

```python
# Drawing a simple rectangle
cv2.rectangle(img = blank_img, pt1 = (10,10), pt2 =(250,250), color = (0,255,0), thickness = 10)
```

```python
cv2.circle(img = blank_img, center = (600, 300), radius = 100, color = (255,0,0), thickness = 8)
```

```python
cv2.putText(img=blank_img, text="Jade", org=(10,980), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale = 2, color = (255,255,255), thickness=2)
```

```python
plt.imshow(blank_img)
```

```python
# Makes the pricture bigger
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
ax.imshow(blank_img)
```
